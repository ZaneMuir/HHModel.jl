### Nernst Potential

driving force
- concentration
- electric potential gradients

$$ E_{ion} = \frac{RT}{zF} \ln \frac{[\mathrm{Ion}]_{out}}{[\mathrm{Ion}]_{in}} $$

- R is the universal gas constant, $8315\ \mathrm{mJ/(K\cdot Mol)}$
- T is temperature in degrees Kelvin ($\text{K} = 273.16 + \text{C}^{\circ}$), body temperature is 310 K.
- F is Farady's constant, $96480\ \mathrm{coulombs/Mol}$
- z is the valence of the ion.

##### approximation:
take $\log_{10}$, using body temprature $T = 310 \text{K}$, for monovalent ions:

$$ E_{\text{ion}} \approx 62 \log \frac{\left[\text{Ion}\right]_{\text{out}}}{\left[\text{Ion}\right]_{\text{in}}}$$

### Ionic Currents and Conductances

$$ I_{ion} = g_{ion} (V - E_{ion}) $$

$g$ is the conductance ($\text{mS}/\text{cm}^2$).

when the conductance is constant, the current is said to be _Ohmic_.
In general, ionic currents in neurons are not Ohmic.

### Equivalent Circuit

according to Kirchhoff' law,
$$ I = C \cdot \frac{dV}{dt} + I_{\mathrm{Na}} + I_{\mathrm{Ca}} + I_{\mathrm{K}} + I_{\mathrm{Cl}} $$

or, as:

$$ C \dot{V} = I - I_{\mathrm{Na}} - I_{\mathrm{Ca}} - I_{\mathrm{K}} - I_{\mathrm{Cl}} $$

when $I = 0$, the membrane potential is typically bounded in the order:

$$ E_{\text{K}} < E_{\text{Cl}} < V_{(\text{at rest})} < E_{\text{Na}} < E_{\text{Ca}} $$
### Resting Potential and Input Resistance

$$ V_{rest} = \frac{g_{Na}E_{Na} + g_{Ca}E_{Ca} + g_{K}E_{K} + g_{Cl}E_{Cl}}{g_{Na} + g_{Ca} + g_{K} + g_{Cl}} $$

the entire euqation can be written in the form:

$$ C \dot{V} = I - g_{inp}(V - V_{rest}) $$

where $ g_{inp} = g_{Na} + g_{Ca} + g_{K} + g_{Cl} $, called _input conductance_.

### Voltage-Clamp and I-V Relation

- instantaneous current-volatge (I-V) relation, $I_0(V_c, V_s)$
    - usu. has a non-monotone N-shape reflecting kinetics
- steady-state current-voltage (I-V) relation, $I_{\infty}(V_s)$

## Conductance

ion gates:

- intracellular agents

$$ I = \bar{g} p (V - E) $$

where p is the average proportion of channels in the open state, $\bar{g}$ is the maximal conductance of the population.

### Voltage-Gated Channels

when thea gating particles are sensitive to the membrane potential, the channels are said to be _voltage-gated_.

$$ p = m^a h^b $$

where $a$ is the number of activation gates and $b$ is the number of inactivation gates per channel.
$m$ can be partially($0 < m <1$) or completely($m=1$) _activated_.
h can be inactivated($h=0$) or deinactivated($h=1$).

such channels do not inactivate, and they result in _**persistent**_ currents.
in contrast, channels that do inactivate result in _**transient**_ currents.

Hodgekin-Huxley gate model of membrane channels

### Activation of Persistent Currents

$$ \dot{m} = \frac{m_{\infty}(V) - m}{\tau(V)} $$

where $m_{\infty}(V)$ is the activation function (sigmoid shape), $\tau(V)$ is the time constant (unimodal shape)

### Inactivation of Transient Currents

$$ \dot{h} = \frac{h_{\infty}(V) - h}{\tau(V)} $$

### Hyperpolarization-Activated Channels

"exceptional" or "weird" channels. $I_{\text{Q}}$(queer), $I_{\text{f}}$(funny), $I_{\text{h}}$(hyperpolarzation-activated), $I_{\mathrm{Kir}}$($\mathrm{K}^+$ inward rectifer)

we can treat them as "normal" transient currents that are always activated ($m = 1$) or deinactivated by hyperpolarization ($h \to 1$)

## The Hudgkin-Huxley Model

$$ C\dot{V} = I - \bar{g}_K n^4 (V-E_K) - \bar{g}_{Na} m^3 h (V - E_{Na}) - g_L (V - E_L) $$

$$ \dot{n} = \alpha_n(V)(1-n) - \beta_n(V) n $$
$$ \dot{m} = \alpha_m(V)(1-m) - \beta_m(V) m $$
$$ \dot{h} = \alpha_h(V)(1-h) - \beta_h(V) h $$

where 

$$ \alpha_n(V) = 0.01 \frac{10 - V}{\exp(\frac{10 - V}{10}) - 1} $$
$$ \beta_n(V) = 0.125 \exp(\frac{-V}{80})$$

$$ \alpha_m(V) = 0.1 \frac{25 - V}{\exp(\frac{25 - V}{10}) - 1} $$
$$ \beta_m(V) = 4 \exp(\frac{-V}{18})$$

$$ \alpha_h(V) = 0.07 \exp(\frac{-V}{20})$$
$$ \beta_h(V) = \frac{1}{\exp(\frac{30 - V}{10}) + 1} $$

parameters for shift resting potential of $V \approx 0$

$$ E_{K}= -12\ \mathrm{mV},\ \ E_{Na}= 120\ \mathrm{mV},\ \ E_{L}= 10.6\ \mathrm{mV}$$

typical maximal conductances are:

$$ \bar{g}_K = 36\ \mathrm{mS/cm^2},\ \ \bar{g}_{Na} = 120\ \mathrm{mS/cm^2},\ \ g_L = 0.3\ \mathrm{mS/cm^2}$$

membrane capacitance is set as $C = 1 \mu\text{F}/\text{cm}^2$ and the applied current as $I = 0 \mu\text{A}/\text{cm}^2$.

### standard form model


$$ C\dot{V} = I - \bar{g}_K n^4 (V-E_K) - \bar{g}_{Na} m^3 h (V - E_{Na}) - g_L (V - E_L) $$

$$ \dot{n} = \frac{n_{\infty}(V) - n}{\tau_n(V)} $$

$$ \dot{m} = \frac{m_{\infty}(V) - m}{\tau_m(V)} $$

$$ \dot{h} = \frac{h_{\infty}(V) - h}{\tau_h(V)} $$

where

$$ n_{\infty} = \frac{\alpha_n}{\alpha_n + \beta_n} $$

$$ \tau_n = \frac{1}{\alpha_n + \beta_n} $$

```julia
alpha_n = (V) -> 0.01 * (10 - V) / (exp(1-V/10) - 1)
beta_n = (V) -> 0.125 * exp(-V/80)
alpha_m = (V) -> 0.1 * (25-V) / (exp(2.5-V/10) - 1)
beta_m = (V) -> 4 * exp(-V/18)
alpha_h = (V) -> 0.07 * exp(-V/20)
beta_h = (V) -> 1 / (exp(3-V/10)+1)
    
n_infty = (V) -> alpha_n(V) / (alpha_n(V) + beta_n(V))
m_infty = (V) -> alpha_m(V) / (alpha_m(V) + beta_m(V))
h_infty = (V) -> alpha_h(V) / (alpha_h(V) + beta_h(V))
    
tau_n = (V) -> 1 / (alpha_n(V) + beta_n(V))
tau_m = (V) -> 1 / (alpha_m(V) + beta_m(V))
tau_h = (V) -> 1 / (alpha_h(V) + beta_h(V))

plot(n_infty, -40, 100, label="\$n_\\infty\$")
plot!(m_infty, -40, 100, label="\$m_\\infty\$")
a = plot!(h_infty, -40, 100, label="\$h_\\infty\$")

plot(tau_n, -40, 100, label="\$\\tau_n\$")
plot!(tau_m, -40, 100, label="\$\\tau_m\$")
b = plot!(tau_h, -40, 100, label="\$\\tau_h\$")

plot(a, b, size=(1200, 600))
```

```julia
"""
du: V, n, m, h, I
"""
function hudgkin_huxley_model(du, u, p, t)
    if 10< t < 11
        u[5] = 8
    elseif 34 < t < 35
        u[5] = 6
    else
        u[5] = 0
    end
    
    
    du[1] = u[5] - 36 * u[2] ^ 4 * (u[1] + 12) - 120 * u[3] ^ 3 * u[4] * (u[1] - 120) - 0.3 * (u[1] - 10.6)
    
    du[2] = (n_infty(u[1]) - u[2]) / tau_n(u[1])
    du[3] = (m_infty(u[1]) - u[3]) / tau_m(u[1])
    du[4] = (h_infty(u[1]) - u[4]) / tau_h(u[1])
    
end

u0 = [0.0, 0.3, 0.05, 0.6, 0.0]
tspan = (0.0, 60.0)

prob = ODEProblem(hudgkin_huxley_model, u0, tspan)
sol = solve(prob, DP5(), reltol=1e-8, abstol=1e-8);
A = plot(sol, vars=(1), ylim=(-20, 120))

# plot(sol, vars=(2),  label="n")
# plot!(sol, vars=(3), label="m")
# a = plot!(sol, vars=(4), label="h", ylim=(0,1))

# g_Na = (t) -> sol(t)[3]^3 * sol(t)[4] * 120
# g_K = (t) -> 36 * sol(t)[2] ^ 4
# plot(g_Na, 0, tspan[2], label="\$g_{Na}\$")
# b = plot!(g_K, 0, tspan[2], label="\$g_{K}\$")

# I_Na = (t) -> g_Na(t) * (sol(t)[1] - 120)
# I_K = (t) -> g_K(t) * (sol(t)[1] + 36)
# I_L = (t) -> 0.3 * (sol(t)[1] - 10.6)
# plot(I_Na, 0, tspan[2], label="\$I_{Na}\$")
# plot!(I_K, 0, tspan[2], label="\$I_{K}\$")
# c = plot!((t) -> I_L(t) + I_Na(t) + I_K(t), 0, tspan[2], label = "\$I_{Na}+I_K+I_L\$")

# B = plot(a,b,c, layout=@layout([a;b;c]), size=(800,600))

# plot(A, B, layout=@layout([A;B]), size=(800,1200), legend=nothing)
```

### Summary of Voltage-Gated Currents

Hodgkin-Huxley gate model

$$ I = \bar{g} m^a h^b (V-E) $$

where
- $I$ - current, ($\mathrm{\mu A/cm^2}$)
- $V$ - membrane voltage, (mV)
- $E$ - reverse potential, (mV)
- $\bar{g}$ - maximal conductance, ($\mathrm{mS/cm^2}$)
- $m$ - probability of activation gate to be open
- $h$ - probability of inactivation gate to be open
- $a$ - the number of activation gates per channel
- $b$ - the number of inactivation gates per channel

$$ \dot{m} = \frac{m_{\infty}(V) - m}{\tau_m(V)} $$

steady-state activation curve $m_{\infty}$ can be approximated by the Boltzmann function:
$$ m_{\infty} (V) = \frac{1}{1 + \exp(\frac{V_{1/2} - V}{k})} $$
where $V_{1/2}$ satisfy $m_{\infty}(V_{1/2}) = 0.5$; 
$k$ is the slope factor, smaller $|k|$ results in steeper $m_{\infty} (V)$

voltage-sensitive time constant $\tau(V)$ can be approximated by the gaussian function:
$$ \tau_m (V) = C_{\mathrm{base}} + C_{\mathrm{amp}} \cdot \exp(\frac{- (V_{\mathrm{max}} - V)^2}{\sigma^2}) $$

where $\tau(V_{\text{max}} \pm \sigma) = C_{\text{base}} + C_{\text{amp}} / e$. 
the gaussian description is often not adequate, so we replace it with other functions whenever appropriate.

---

```julia
function activation(Vhalf, k)
    return (V) -> 1 / (1+exp((Vhalf-V)/k))
end

function tau(Vmax, sigma, Camp, Cbase)
    return (V) -> Cbase + Camp * exp(-((Vmax - V)/sigma)^2)
end

n_infty = activation(12, 15)
tau_n = tau(-14, 50, 4.7, 1.1)

m_infty = activation(25, 9)
tau_m = tau(27, 30, 0.46, 0.04)

h_infty = activation(3, -7)
tau_h = tau(-2, 20, 7.4, 1.2)
```

```julia
plot(n_infty, -40, 100)
plot!(m_infty, -40, 100)
a = plot!(h_infty, -40, 100)

plot(tau_n, -40, 100)
plot!(tau_m, -40, 100)
b = plot!(tau_h, -40, 100)

plot(a,b, size=(1200, 600))
```
